```python
import numpy as np
import pandas as pd
import os
os.chdir('../')
from root_dir import * 
import matplotlib as mpl
import matplotlib.pyplot as plt
from distutils.spawn import find_executable
%matplotlib inline
```

```python
for y in [0,1]:
    for w in [1,2]:
        plot_feature = pd.read_csv('%s/privateDemographics/results/toy/outliers/W%dball_feature_y_%d.csv' % (root_dir, w, y)) 
        plot_grad = pd.read_csv('%s/privateDemographics/results/toy/outliers/W%dball_grad_y_%d.csv' % (root_dir, w, y)) 

        plot = pd.DataFrame({
            'weight': plot_grad.weight, 
            'feature': plot_feature.outlier / plot_feature.regular, 
            'grad': plot_grad.outlier / plot_grad.regular, 
        })

        ax = plot.plot('weight', 'feature')
        plot.plot('weight', 'grad', ax = ax)
        ax.set_title('W%d distance, y=%d' % (w, y))
```

```python
w=1

width = 8
height = 8

# check whether latex is available
if find_executable('latex'):
    plt.rc('text', usetex=True)
    plt.rc('font', family='serif', serif='times new roman')
plt.rc('xtick', labelsize=28)
plt.rc('ytick', labelsize=28)
plt.rc('axes', labelsize=23)
plt.rc('axes', linewidth=2)
mpl.rcParams['patch.linewidth']=2 #width of the boundary of legend

fig, ax = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True) #plot two subgraphs in one figure, 1 by 2 size, sharing y-axis
fig.subplots_adjust(left=.15, bottom=.2, right=0.99, top=0.97, wspace=0.02) #margin of the figure
fig.set_size_inches(width, height) #exact size of the figure


for y in [0,1]:
    plot_feature_outlier = pd.read_csv('%s/privateDemographics/results/toy/outliers/W%dball_feature_y_%d.csv' % (root_dir, w, y)) 
    plot_grad_outlier = pd.read_csv('%s/privateDemographics/results/toy/outliers/W%dball_grad_y_%d.csv' % (root_dir, w, y)) 
    plot_feature_minority = pd.read_csv('%s/privateDemographics/results/toy/minority/W%dball_feature_y_%d.csv' % (root_dir, w, y)) 
    plot_grad_minority = pd.read_csv('%s/privateDemographics/results/toy/minority/W%dball_grad_y_%d.csv' % (root_dir, w, y)) 
    ax[y,0].plot(plot_feature_outlier.weight, plot_feature_outlier.outlier / plot_feature_outlier.regular, label = 'feature', marker = '<', linewidth=2)
    ax[y,0].plot(plot_feature_outlier.weight, plot_grad_outlier.outlier / plot_grad_outlier.regular, label = 'gradient', marker = '.', linewidth=2)
    ax[y,1].plot(plot_feature_minority.weight, plot_feature_minority.minority / plot_feature_minority.other, label = 'feature', marker = '<', linewidth=2)
    ax[y,1].plot(plot_feature_minority.weight, plot_grad_minority.minority/ plot_grad_minority.other, label = 'gradient', marker = '.', linewidth=2)

ax[0,0].set_title('outlier', fontsize = 28)
ax[0,1].set_title('minority', fontsize = 28)
ax[1,0].set_ylabel('y=1')
ax[0,0].set_ylabel('y=0')
ax[0,1].legend(fontsize=20)
```
